```python
import matplotlib.pyplot as plt

# Step 1: Input number of subjects
# Make sure to enter an integer number here, not text.
num_subjects = int(input("Enter number of subjects: "))

subjects = []
marks = []

# Step 2: Take subject names and marks
for i in range(num_subjects):
    subject = input(f"Enter name of subject {i+1}: ")
    mark = float(input(f"Enter marks obtained in {subject}: "))
    subjects.append(subject)
    marks.append(mark)

# Step 3: Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(subjects, marks, color='skyblue')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Student Marks Visualization')
plt.ylim(0, 100)  # Assuming max marks = 100
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Step 4: Show the plot
plt.tight_layout()
plt.show()
```
